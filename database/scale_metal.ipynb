# Construct some example species with thermo

```python
import rmgpy.data.rmg
database = rmgpy.data.rmg.RMGDatabase()
database.load(
    path = rmgpy.settings['database.directory'],
    thermo_libraries = ['primaryThermoLibrary', 'surfaceThermoPt111'],
    transport_libraries = [],
    reaction_libraries = [],
    seed_mechanisms = [],
    kinetics_families = 'none',
    kinetics_depositories = ['training'],
    depository = False,
) 

```

```python
HX = rmgpy.species.Species(smiles='[H]*')
display(HX)
```

```python
database.thermo.get_thermo_data(HX, metal_to_scale_to='Cu111')
```

```python
thermo0 = database.thermo.get_thermo_data(HX)
```

```python
database.thermo.correct_binding_energy(thermo0, HX, metal_to_scale_to='Cu111')
```

```python
database.thermo.correct_binding_energy(thermo0, HX, metal_to_scale_to='Cu111').to_nasa(Tmin=300, Tmax=1500, Tint=1000)
```
